* kind (O Tipo do Gráfico): Diz ao Pandas como os dados devem ser desenhados.

* 'barh': Barras horizontais (ótimo para categorias com nomes longos).

* 'bar': Barras verticais padrão.

* 'line': Gráfico de linha (padrão se você não colocar nada).

* 'pie': Gráfico de pizza/tortas.

* figsize (O Tamanho da Tela): Controla as dimensões físicas da imagem em polegadas: (LARGURA, ALTURA). Aumente os números se o gráfico tiver muitos elementos ou se as legendas estiverem se atropelando.

* color (A Cor dos Elementos): Define a cor das barras, linhas ou fatias. Você pode passar o nome em inglês ('blue', 'red', 'green') ou códigos hexadecimais de paletas que achar na internet (ex: '#FFD700' para dourado).

* ; (O Ponto e Vírgula no final da linha): Não é obrigatório, mas serve para uma "limpeza visual" no Jupyter/Colab. Se você tirar o ;, o Python vai imprimir um texto técnico chato em cima do gráfico (algo como <Axes: xlabel='...'>). O ponto e vírgula esconde esse texto e mostra apenas o gráfico limpo.

```python
df_preco_tipo.plot(kind='barh', figsize=(14,10), color = 'blue');
```

separo os tipos em dois, um de imoveis comerciais e o outro de imoveis residenciais

```python
Imoveis_comerciais = ['Conjunto Comercial/Sala', 'Loja/Salão', 'Galpão/Depósito/Armazém', 'Casa Comercial', 'Loja Shopping/ Ct Comercial', 'Box/Garagem', 'Prédio Inteiro', 'Indústria', 'Terreno Padrão', 'Chácara', 'Loteamento/Condomínio', 'Hotel']
print(Imoveis_comerciais)
```

```python
dados.query('@Imoveis_comerciais in Tipo')
```

```python
imoveis_comerciais = dados.query('@Imoveis_comerciais in Tipo')
```

```python
imoveis_residenciais = dados.query('@Imoveis_comerciais not in Tipo')
```

```python
imoveis_residenciais
```

aqui crio outras duas variaveis agrupando apenas as colunas por tipo e mostrando apenas a coluna valor

```python
df_imoveis_residenciais = imoveis_residenciais.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

```python
df_imoveis_residenciais.plot(kind='barh', figsize=(14,10), color = 'blue');
```

aqui mostro todos os nomes dentro dos resienciais apenas na coluna tipo

```python
imoveis_residenciais.Tipo.unique()
```

aqui conto quantas vezes cada nome se repete dentro da coluna

```python
imoveis_residenciais.Tipo.value_counts().to_frame().sort_values('Tipo')
```

aqui converto quanto cada uma representa em % do total dentro da coluna tipo

```python
imoveis_residenciais.Tipo.value_counts(normalize=True)
```

```python
imoveis_residenciais.Tipo.value_counts(normalize=True).to_frame().sort_values('Tipo')
```

```python

df_percentual = imoveis_residenciais.Tipo.value_counts(normalize=True).to_frame().sort_values('proportion', ascending=False)
```

```python
df_percentual.plot(kind='bar', figsize=(14,10), color = 'blue', xlabel='Tipo', ylabel='Percentual');
```
